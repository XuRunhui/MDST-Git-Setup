```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a pair plot for multiple variables
df = pd.read_csv("../data/starbucks.csv")
sns.pairplot(data=df, vars=['calories', 'iron', 'total fat'], hue='beverage_category')
plt.title('Pair Plot')
plt.show()
plt.tight_layout()
```
